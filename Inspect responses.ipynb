```python
import pandas as pd
from ast import literal_eval
from matplotlib import pyplot as plt
from collections import Counter
from collections import defaultdict
import numpy as np
```

```python
def co_occurrence(responses_list):
    
    window_size = 20 # just a random high number: it will go until the end of the response in our case
    d = defaultdict(int) #it sets an empty dictionary
    vocab = set()
    
    for response in responses_list:
        response = response
        for i in range(len(response)):
            word = response[i] #takes a word
            vocab.add(word)  # adds to vocab
            next_words = response[i+1 : i+1+window_size] #takes all the folowing ones
            for word2 in next_words: #iterates over the following words and adds a +1
                key = tuple(sorted([word2, word]))
                d[key] += 1 #it puts the number in the dictionary, at the key "(word, word2)" or "(word2, word)", depending on the alphabetical order.

    # transforms the dictionary into dataframe
    vocab = sorted(vocab)
    df = pd.DataFrame(data=np.zeros((len(vocab), len(vocab)), dtype=np.int16), index=vocab, columns=vocab)

    serie = pd.Series(d).reset_index()
    serie.columns = 'word1', 'word2', 'co-occurrences'
    
    return serie
```

### Co-occurrences matrix

### People

```python
responses = data.loc[data['domain'] == 'people', 'responses_v2'].apply(literal_eval)

resp_words_tmp = [i.keys() for i in responses]

resp_words = []
for i in resp_words_tmp:
    tmp = []
    for j in i:
        tmp.append(j)
    resp_words.append(tmp)
    
    
resp_values_tmp = [i.values() for i in responses]

resp_values = []
for i in resp_values_tmp:
    tmp = []
    for j in i:
        tmp.append(j)
    resp_values.append(tmp)
    
    
keys_tmp = []
for i in resp_words:
    for j in i:
        keys_tmp.append(j)
        
keys = set(keys_tmp)
```

```python
cooc = co_occurrence(resp_words)
cooc = cooc.sort_values(by='co-occurrences', ascending = False)
cooc.head()
```
